```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
# Converts a string of dictionaries into a list of names.
```

```python
def get_director(text):
    """
    This function fetches the director of a movie from a string of JSON data.

    Args:
        text (str): A string of JSON data.

    Returns:
        list: A list of directors.
    """

    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
movies['crew'] = movies['crew'].apply(get_director)
#This code updates the crew column to contain the director of each movie.
#The apply() method is used to apply the get_director() function to each element of the crew column.
#The get_director() function fetches the director of a movie from a string of JSON data.
```

```python
def collapse(L):
    """
    This function collapses a list of strings into a single string.

    Args:
        L (list): A list of strings.

    Returns:
        str: A single string.
    """

    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return "".join(L1)
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
#This code collapses the cast, crew, genres, and keywords columns into single strings.
#The apply() method is used to apply the collapse() function to each element of the respective columns.
#The collapse() function collapses a list of strings into a single string
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())

#This code splits the overview column into a list of words.
#The apply() method is used to apply the lambda function to each element of the overview column.
#The lambda function splits a string into a list of words.

```

```python

movies['tags'] = movies['overview'].apply(lambda x: ' '.join(x)) + movies['genres'].apply(lambda x: ' '.join(x)) + movies['keywords'].apply(lambda x: ' '.join(x)) + movies['cast'].apply(lambda x: ' '.join(x)) + movies['crew'].apply(lambda x: ' '.join(x))
movies['tags'] = movies['tags'].apply(lambda x: x.split())

#This code creates a new column called tags that contains a combination of the overview, genres, keywords, cast, and crew columns.
#The + operator is used to combine the strings in the respective columns.
#The tags column can be used to generate recommendations for movies based on the user's interests.
```

```python
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
#This code drops the overview, genres, keywords, cast, and crew columns from the movies DataFrame.
#The drop() method is used to drop columns from a DataFrame.
#The columns argument specifies the columns that should be dropped.

new['tags'] = new['tags'].apply(lambda x: " ".join(x))
# This code joins the words in the `tags` column with spaces.
# The `apply()` method is used to apply the `lambda` function to each element of the `tags` column.
# The `lambda` function joins the words in a list with spaces.
```

```python
new.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
# Create a CountVectorizer object

vector = cv.fit_transform(new['tags']).toarray()
# Convert the tags column into a vector of features.
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
new[new['title'] == 'The Lego Movie'].index[0]
#The iloc method of a DataFrame is used to access a row or column by its index. The iloc[0] argument to the iloc method specifies that the first row of the DataFrame should be returned.
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distance = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distance[1:6]:
        print(new.iloc[i[0]].title)
        
#The recommend() function takes a movie as an argument and returns a list of movies that are similar to the movie. The function works as follows:

#The function gets the index of the movie in the new DataFrame.
#The function calculates the cosine similarity between the movie and all other movies in the new DataFrame.
#The function sorts the cosine similarity scores in descending order.
#The function prints the titles of the top 5 movies with the highest cosine similarity scores.
        
    
```

```python
recommend('The Lego Movie')
```
